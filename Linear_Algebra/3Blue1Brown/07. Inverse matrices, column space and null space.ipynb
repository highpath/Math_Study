# 1. Linear system of equations
## 1) Shape
- each variable scaled by some constant
- added to each other
- ex) $2x+5y+3z=-3$

## 2) Organize
1. Throw variables on the left
2. Lingering constants on the right
3. Vertically align variables
4. Add zeros as needed

## 3) Matrix vector product
- Matrix(Coefficients) x Vector(Variables) = Vector(Constants)
- $A \vec{x}=\vec{v} $
- A: Transformation

# 2. Solutions

## 1) $det(A)\not=0$ case
- There's a single, unique solution.
- $A^{-1}$ exists

1. Inverse Transformation
    - transformation in reverse
    - $A^{-1}A=I$
2. Identity Transformation
    - The transformation that does nothing

## 2) $det(A)=0$ case
- No function do transformation in reverse
- Solutions can still exist when $det(A)=0$

# 3. Rank

- The number of dimensions in the output of a transformation
- The number of dimensions in the column space
- full rank: rank as high as it can be / number of columns

# 4. Column space
- Set of all possible outputs $A\vec{v}$ = "Column space" of $A$
- The span of the columns of the matrix
- zero vector always be included in the column space

# 5. Null space or Kernel
- the space of all vectors that become null
- $A\vec{x}=\begin{bmatrix}0\\0\end{bmatrix}$
