### How would you describe one of these(Linear transformations) numerically?

- It can be represented by the two coordinates for where $\vec{i}$ lands and the two coordinates for where $\vec{j}$ lands.

- It's common to package these coordinates into a two-by-two grid of numbers. (2x2 Matrix)

$\vec{v} = a \hat{i}+b \hat{j}$

Transformed $\vec{v} = a($Transformed $ \hat{i}) + b($Transformed $ \hat{j})$

$$ 
\begin{bmatrix} a&b \\ c&d\end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix}
= x \begin{bmatrix} a \\ c\end{bmatrix} + y \begin{bmatrix} b \\ d \end{bmatrix}
= \begin{bmatrix} ax+by \\ cx+dy \end{bmatrix}
$$
