### Think of each coordinate as a **scalar** (how each one scales vectors)

# 1. Basis vectors

- **basis vectors** of $xy$ coordinate system: 
$\hat{i}= \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \ \hat{j}=\begin{bmatrix} 0 \\ 1 \end{bmatrix}$
- think of the vector coordinates of a vector as scalars that scale $\hat{i}$ and $\hat{j}$
- $\begin{bmatrix}a \\ b\end{bmatrix} = a\hat{i}+b\hat{j}$

### What if we chose different basis vectors?

- we can choose different basis vectors, and get a completely reasonable, new coordinate system
- any time we describe vectors numerically, it depends on an implicit choice of what basis we're using.

# 2. Linear combination
- Scaling two vectors and adding them
- $a\vec{v}+b\vec{w}$ (a, b = Scalars)

# 3. Span
- the set of all their linear combinations
- $a\vec{v}+b\vec{w}$ (Let $a$ and $b$ vary over all real numbers)
- what are all the possible vectors you can reach using only **two fundamental operations**.

# 4. Vectors vs Points

- if dealing with a collection of vectors, it's convenient to think of sets of vectors as points
- The point at the tip of that vector with its tail on the origin
- ex) if every possible vector whose tip sits on a certain line, just think about the line itself.

- **Think of individual vectors as arrows**
- **Think of sets of vectors as points**

# 5. Linearly Dependent or Independent

- Linearly Dependent
    - If a vector can be represented by linear combination of other vectors
    - $\vec{u} = a\vec{v} + b\vec{w}$ (For some values of a and b)
- Linearly Independent
    - $\vec{u} \not= a\vec{v} + b\vec{w}$ (For all values of a and b)

# 6. Technical definition of basis
- The **basis** of a vector space is a set of **linearly independent** vectors that **span** the full space
